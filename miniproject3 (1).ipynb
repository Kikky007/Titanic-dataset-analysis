```python
# import important libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# Load the dataset
df = pd.read_csv("titanicdataset11.csv")

# Show first few rows
print(df.head(10))
```

```python

# Checking for missing values
print(df.isnull().sum())
```

```python
# Calculate the median age
median_age = df['age'].median()

# Print the result
print("Median Age:", median_age)
```

# Data Cleaning

```python
# Fill missing 'age' values with the median
df['age'] = df['age'].fillna(df['age'].median())

# Remove the 'deck' column
df.drop(columns=['deck'], inplace=True)

# Drop rows where 'embarked' is missing
df = df.dropna(subset=['embarked'])

# Preview the cleaned data
print(df.head())
```

# Business Questions for Analysis

```python
# Question 1: What is the survival rate of passengers?

survival_rate = df['survived'].mean()
print(f"Survival Rate: {survival_rate:.2%}")
```

```python
# 2. What is the gender distribution of passengers?
df['sex'].value_counts().plot(kind='bar', title='Gender Distribution', color=['purple', 'green'])
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

**This barchart indicates that male passengers are more than female passengers.**

```python
# 3. How does survival rate differ by class?
sns.barplot(x='pclass', y='survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

**Analysis: Survival Rate by Passenger Class**

**This bar chart reveals that:**

**1st Class passengers had the highest survival rate.**

**2nd Class passengers had a moderate survival rate.**

**3rd Class passengers experienced the lowest survival rate.**

**This descending trend indicates that socio-economic status played a significant role in survival outcomes. Higher-class passengers may have had better access to lifeboats, quicker evacuation routes, or received priority during rescue efforts, reflecting the socio-economic inequalities present at the time.**

**This insight emphasizes how class-based distinctions can have life-or-death consequences in crisis situations.**

```python
# 4. What is the distribution of passenger ages?
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate skewness of the age distribution
age_skewness = df['age'].skew()
age_skewness

# The age skewness is 0.508, which is greater than 0, 
# which shows that the distribution has a longer tail on the right side which makes it right skewed
```

**NB:** I filled the missing values in the age column(n = 177) with the median age, so there are some bias there.

**This histogram shows the age distribution is right-skewed, meaning there are more younger passengers with a gradual decline in numbers as age increased. The age with the highest frequency is approximately 28, followed by the mid-twenties, then early twenties. Overall, the Titanic seemed to have a large frequency of people in their 20s and 30s.**



```python
# 5. How many passengers embarked from each location?
df['embarked'].value_counts().plot(kind='bar', title='Passengers by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Passenger Count')
plt.show()

# Calculate the percentage of the passengers embarked from each location

# Count the number of passengers by embarkation point
embarked_counts = df['embarked'].value_counts()

# Convert to percentages
embarked_percentages = (embarked_counts / embarked_counts.sum()) * 100

# Round to 2 decimal places for readability
embarked_percentages = embarked_percentages.round(2)

print(embarked_percentages)
```


**The majority of passengers (approximately 72%) embarked from Southampton (S), which was the Titanic’s departure point in England.**

**Cherbourg (C) was the next most common embarkation point, accounting for about 19% of the passengers.**

**Queenstown (Q) had the lowest number, contributing around 9% of the total passengers.**

**This barchart shows that Southampton served as the main boarding port for Titanic passengers, likely due to its status as a major British port city.**

```python
# 6. What is the average fare paid per class?
avg_fare_per_class = df.groupby('pclass')['fare'].mean()
print("\nAverage Fare Paid Per Class:")
print(avg_fare_per_class)
```

```python
# 7. How does gender affect survival rates?
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

**This barchart indicates that females had a significantly higher survival rate compared to males.**
This shows it is likely that the "women and children first" policy followed during the evacuation.
 

```python
# 8. Correlation between fare and survival?
sns.boxplot(x='survived', y='fare', data=df)
plt.title('Fare vs Survival')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()

#calculate the correlation betwen fare and survival
correlation = df['fare'].corr(df['survived'])
print('The correlation between Fare and Survival:', correlation)
```

**The boxplot shows that the passengers who survived generally paid higher fares compared to those who survived.**
The correlation between fare and survival is approximately 0.255. This indicates a positive but weak correlation, this means that those that paid higher fare had a slightly survival rate.  

```python
# 9. How does number of siblings/spouses aboard affect survival?
sns.barplot(x='sibsp', y='survived', data=df)
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.xlabel('Siblings/Spouses Aboard')
plt.ylabel('Survival Rate')
plt.show()
```

**This bar chart shows that passengers with 1 siblings/spouses had the highest survival rate followed by 2 sibling/spouses.** 
Those traveling alone (0 sibsp) or with larger families (≥3) had lower chances of survival, with those traveling alone higher than those with 3 sibsp and those with 4 sibsp the lowest. This may be due to the difficulty of evacuating larger groups together.


```python
# 10. What is the age distribution by class?
sns.boxplot(x='pclass', y='age', data=df)
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```


**This boxplot reveals that there was higher proportion of adults(late twenties and late forties) in the first class, and more of young adults in the third class, which suggests that socioeconomic factors may have influenced who traveled in which class, with wealthier, older individuals more likely in higher classes.**

```python
# 11. Survival rate by embarkation point
sns.barplot(x='embarked', y='survived', data=df)
plt.title('Survival Rate by Embarkation Point')
plt.xlabel('Embarkation')
plt.ylabel('Survival Rate')
plt.show()
```

**This bar chart shows that passengers who boarded from Cherbourg (C) had a slightly higher chance of survival.
Those who embarked from Queenstown (Q) were more likely not to survive.
passengers from Southampton (S) also had the lowest survival rate. This may likely be due to many first class passengers boarded from Cherbourg**

```python
# 12. Impact of having parents/children aboard on survival
sns.barplot(x='parch', y='survived', data=df)
plt.title('Survival Rate by Parents/Children Aboard')
plt.xlabel('Parents/Children Aboard')
plt.ylabel('Survival Rate')
plt.show()
```

**This bar Chart shows that passengers with 1, 2 and 3 parch were more likely to survive, and passengers alone were likely not to, with passengers with 5 parch, the lowest survival rate, which may be due to the difficulty of coordinating escape for larger groups during the chaos.**
